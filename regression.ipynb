## Úloha 2
### Vyberte si jednu numerickou proměnnou jako regresor.

Natrénujme lineární regresní model, který bude vysvětlovat jakým způsobem proměnná `Phtolyas` ovlivňuje proměnnou `Percent`.

```r
fit.pht.lin=lm(Percent~Phtolyas)
summary(fit.pht.lin)
```

Funkce `summary()` nám vypsala několik důležitých věcí:
* Lineární regrese: `Percent = 45.58 - 5.65 * Phtolyas`. Konkrétně vztah říká, že pro každý jeden jednotkový nárůst hodnoty proměnné `Phtolyas` dojde k průměrnému snížení `Percent` o 5,65 procentních bodů. Intercept představuje hodnotu podílu nevylíhnutých vajec žab za podmínky, že hodnota `Phtolyas` je rovná nule.
* Medián reziduí je trochu menší než nula, ale blizko ní, což nám dává důvod věřit, že naše rezidua jsou nevychýlená (musíme však ověřit testem).
* Intercept a regresor `Phtolyas` jsou signifikantní v tomto modelu.
* Koeficient determinace je roven 0.6, adjustovaný koeficient determinace - taky 0.6.

Zkusíme taky ověřit nelineární vztah mezi regresorem a regresandem (na grafu vypadá, že polynomiální regrese by mohla být vhodnější).

```r
fit.pht.sq=lm(Percent~I(Phtolyas^2) + Phtolyas)
summary(fit.pht.sq)
```

* Kvadratická regrese: `Percent = 24.55 + 9.26 * Phtolyas - 1.63 * Phtolyas^2`. Intercept nám zase ukazuje hodnotu podílu nevylíhnutých vajec žab za podmínky, že hodnota `Phtolyas` je rovná nule. Dále v modelu figuruje lineární koeficient 9.26 a kvadratický koeficient -1.63, což znamená, že při postupném zvyšování hodnoty `Phtolyas` (začínající od nuly) se hodnota `Percentu` zvyšuje, ale poté se růst `Percentu` zpomaluje a dokonce začíná klesat.
* Medián reziduí je ještě bližší k nule, než v předchozím případě, což nám dává důvod věřit, že naše rezidua jsou nevychýlená (musíme však ověřit testem).
* Intercept a regresor $Phtolyas^{\text{2}}$ jsou signifikantní v tomto modelu.
* Nezamítáme nulovou hypotézu o tom, že regresor `Phtolyas` nemá vliv na regresand `Percent` v tomto modelu na standardní hladině významnosti.
* Koeficient determinace je roven 0.65, adjustovaný koeficient determinace = 0.639, oba koeficienty jsou vyšší než v předchozím modelu.

Ověříme také případný exponenciální vztah.

```r
fit.pht.exp=lm(Percent~I((exp(Phtolyas))))
summary(fit.pht.exp)
```

* Exponenciální regrese: `Percent = 35.74 - 0.019 * Exp(Phtolyas)`. Intercept má stejný význam jako dříve. Tato závislost nám říká, že když se hodnota `Phtolyas` zvětšuje, hodnota `Percent` se pomalu snižuje, ale s rostoucí rychlostí (čím víc se hodnota regresoru zvyšuje, tím rychleji klesá hodnota regresandu). 
* Medián reziduí je nejbližší k nule ze všech modelů, což nám dává důvod věřit, že naše rezidua jsou nevychýlená (musíme však ověřit testem).
* Intercept a regresor `Exp(Phtolyas)` jsou signifikantní v tomto modelu.
* Koeficient determinace je roven 0.64, adjustovaný koeficient determinace = 0.638.

Všechny modely mají téměř stejnou hodnotu koeficientu determinace nad 0,6, což není příliš vysoké číslo. To znamená, že všechny naše modely nejsou moc přesné. Nicméně nejvyšší koeficient determinace má kvadratická regrese, takže ji budeme použivat dále.

Zobrazíme kvadratický vztah graficky.

```r
ggplot(ex2414, aes(x=Phtolyas, y=Percent)) +
    geom_point() +
    stat_smooth(method="lm", formula=y~x+I(x^2), col="blue", se=F) +
    stat_smooth(method="lm", formula=y~x+I(x^2)) +
    labs(title = "Quadratic relationship between Percent and Phtolyas")+
    theme(plot.title = element_text(size = 20, hjust = 0.5))
```

```r
summary(fit.pht.sq$resid)
hist(fit.pht.sq$resid, prob=T, col="Lightblue", main='Histogram reziduí kvadratického modelu')
plot(fit.pht.sq, which=1, col=c("blue"))
```

Histogram reziduí vypadá jako normálně rozdělený (to však tvrdit nemůžeme). Na grafu reiduji body jsou rozmištěný celkem symetrický vůči 0 (to však taky tvrdit nemůžeme).

Intercept v tomto modelu udává hodnotu `Percentu`, když se jedná o lokaci ThreeCreeks. Jedná se tedy o průměrnou hodnotu `Percentu` (na datech, která máme) pro tuto lokaci. * Všecny hodnoty kategoriální proměnné `Location` jsou signifikantní v tomto modelu.
* Koeficient determinace je roven 0.47, adjustovaný koeficient determinace = 0.45. Kvailta modelu je horší než kvalita předchozích modelů (podle koeficientu determinace), takže proměnná Location hůř vysvětluje podíl nevylíhnutých vajec žab.

```r
aov_result <- aov(fit.loc)
aov_result
anova(aov_result)
```

Na obvyklé hladině významnosti 𝛼 = 5% zamítneme nulovou hypotézu ve prospěch alternativy, že je střední hodnota `Percentu` různá pro každou lokaci. To taky znamená, že v této formuli `celková variabilita = variabilita uvnitř výběrů + variabilita mezi výběry` je variabilita mezi výběry významně velká a hypotézu shody rozptylu zamítneme. Z praktického pohledu to znamená, že hodnoty proměnné 'Percent' se mezi jednotlivými lokacemi různí a že proměnná `Location` má vliv na proměnnou `Percent`.

## Úloha 5
### Uvažujte model s předchozími dvěma regresory a alespoň jedním dalším ze zkoumaného datového souboru.
> Prozkoumejte, které z regresorů jsou pro model významné a které ne.

Přidáme další interakci s regresorem `Treat` a nafitujeme lineární model.

```r
fit.all <- lm(Percent~(I(Phtolyas^2) + Phtolyas)*Location*Treat, data = df)
summary(fit.all)
```

Nejprve si všimneme, že se zvýšil adjustovaný koeficient determinace. Také vidíme, že významné koeficienty jsou zejména interakce mezi `Location` a `Treat`.


>Pomocí vhodných nástrojů se pokuste najít vhodný finální model, který chování zkoumané proměnné dobře vysvětluje, ale neobsahuje nevýznamné komponenty.


Použijeme funkci `step()`, která na základě Akaikeho informačního kritéria sestaví nejvhodnější model pro lineární regresi.

```r
step(lm(Percent~1), scope=list(lower=~1,upper=~Location*Phtolyas*Treat))
```

Dostali jsme nejlepší model podle AIC, nyní bychom měli provést jeho fitting a podívat se na výsledky.


```r
fit.step = lm(formula = Percent ~ Phtolyas + Treat + Location + Phtolyas:Treat)
summary(fit.step)
```

Z toho vidíme, že tento model má téměř stejný adjustovaný koeficient determinace jako model s interakcemi všech příznaků, a navíc máme v tomto modelu větší procento významných parametrů.


Porovnáme modely příkazem `anova()` pomocí `F-testu`.


```r
anova(fit.step, fit.all)
```


Na standardní hladině významnosti 5% nebyl nalezen statisticky významný rozdíl mezi modely, a proto preferujeme ten jednodušší (fit.step).

Víme, že kvadrát regresoru Phtolyas může lépe charakterizovat vysvětlovanou proměnnou, takže bychom mohli zahrnout tento regresor do modelu a tím ho vylepšit.

```r
fit.step1 = lm(formula = Percent ~ I(Phtolyas^2) + Phtolyas + Treat + Location + (I(Phtolyas^2) + Phtolyas):Treat)
summary(fit.step1)
```

Vidíme, že adjustovaný koeficient determinace se zlepšil. Nyní bychom měli otestovat, zda se jedná o statisticky významné zlepšení.

```r
anova(fit.step, fit.step1)
```

Závěr: Na standardní hladině významnosti 5% **nebyl nalezen statisticky významný rozdíl mezi modely**. Z tohoto důvodu preferujeme ten jednodušší (fit.step).

> Výsledný model prakticky interpretujte.

```r
summary(fit.step)
```

Máme výsledný adjustovaný koeficient determinace rovný 0.86, což je blízko hodnotě 1. To znamená, že model dobře popisuje varianci závislé proměnné. Do modelu byly zahrnuty všechny příznaky, což naznačuje, že jsou všechny příznaky v našem datasetu užitečné. Model není příliš složitý, protože obsahuje pouze jednu interakci, která naznačuje, že výroba phtolyasu je silně závislá na UV záření.

## Úloha 6
### U finálního modelu pomocí vhodných testů ověřte předpoklady metod, které jste použili.

Předpoklady, které musíme ověřit: 
* normalitu odchylek 
* homoskedasticitu reziduí
* nekorelovanost reziduí

Začneme s ověření normality odchylek. Pro tento účel použijeme **Shapiro-Wilkův** test normality.

```r
plot(fit.step, which=2, col=c("blue"))
shapiro.test(fit.step$residuals)
```

Na hladině vyznamnosti 5% dostaváme, že **nejsou splněny předpoklady normality odchylek**. Z toho vyplyvá, že **F statistika**, kterou jsme použivali v `anova()`, **nemá Fisherovo-Snedecorovo F-rozdělení** s  $k-r$ a $n-k-1$ stupni volnosti. To znamená. že **p hodnota**, která by měla vyjadřovat pravděpodobnost, že F-rozdělení bude vyšší než spočítaná F-statistika, **není příliš spolehlivá**. 

Můžeme zkusit porovnat model a podmodel pomocí **testu poměru věrohodnosti**, abychom zjistili, zda je věrohodnost širšího modelu významně vyšší. Je však třeba mít na paměti, že i tento test vyžaduje předpoklady normality odchylek.

```r
anova(fit.step, fit.step1, test="Chisq")
```

Vidíme, že v tomto případě mezi modely na hladině významnosti 5% **není statisticky významný rozdíl**. Z tohoto důvodu preferujeme ten jednodušší (fit.step).

Teď se podíváme na homoskedasticitu reziduí. Pro tento účel použijeme **Breuschův-Paganův** test.

```r
plot(fit.step, which=3, col=c("blue"))
ols_test_breusch_pagan(fit.step)
```

Závěr: 
 
P-hodnota je **menší** než 5%, proto **zamítáme** nulovou hypotézu ($\mathbf{H_0:}$ **rezidua jsou homoskedastická**) ve prospěch alternativy, že jsou heteroskedastická, což znamená, že rozptyl reziduí není konstantní. V takovém případě bychom měli uvažovat nad vhodnou transformací dat, která by stabilizovala rozptyl.

Teď se podíváme na korelaci reziduí. Pro tento účel použijeme **Breuschův-Godfreyúv** test namísto Durbinova-Watsonova testu, protože Breusch-Godfreyův test [nepotřebuje](http://math.furman.edu/~dcs/courses/math47/R/library/lmtest/html/bgtest.html) předpoklad normality reziduí.

```r
plot(fit.step, which=1, col=c("blue"))
bgtest(fit.step)
```

Závěr:

P-hodnota je ale **větší** než 5%, proto **nezamitáme** nulovou hypotézu ($\mathbf{H_0:}$ **Rezidua jsou nekorelovaná**). Předpoklad nekorelovaností je splněn. To znamená, že model byl schopen dobře zachytit vzorce v datech a že v reziduích nejsou žádné významné zbývající vzorce nebo struktury. To je obecně považováno za dobrou věc.
